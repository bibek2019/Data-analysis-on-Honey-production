# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python
selected_data=['numcol','yieldpercol','totalprod','stocks','priceperlb','prodvalue']
correlation = data[selected_data].corr()
sns.heatmap(correlation, annot=True)
```

Positive Correlation: A robust positive correlation is evident between the number of honey-producing colonies (numcol) and total honey production (totalprod), with a correlation coefficient of 0.953594. This indicates a significant tendency for total honey production to increase as the number of colonies rises. Similarly, there is a strong positive correlation between total honey production (totalprod) and honey stocks (stocks), reflected by a correlation coefficient of 0.878830. This suggests that higher total production is associated with elevated stocks of honey.

Negative Correlation: A moderate negative correlation exists between the price per pound of honey (priceperlb) and the number of honey-producing colonies (numcol), as indicated by a correlation coefficient of -0.232701. This implies that an increase in the number of colonies tends to correspond with a decrease in the price per pound of honey.

Other Correlations: Weaker correlations are observed among other pairs of variables.
